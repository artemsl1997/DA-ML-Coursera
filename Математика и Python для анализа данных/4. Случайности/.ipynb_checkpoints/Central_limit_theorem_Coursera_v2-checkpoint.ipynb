## Центральная предельная теорема своими руками
# Проверка работы центральной предельной теоремы на выбранном непрерывном распределении
# Выбранное распределение: Функция распределения Лапласа
https://ru.wikipedia.org/wiki/%D0%A0%D0%B0%D1%81%D0%BF%D1%80%D0%B5%D0%B4%D0%B5%D0%BB%D0%B5%D0%BD%D0%B8%D0%B5_%D0%9B%D0%B0%D0%BF%D0%BB%D0%B0%D1%81%D0%B0 !image.png

```python
# импорт необходимых библиотек
import numpy as np
from scipy.stats import laplace, norm
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# параметры распределения Лапласа
beta = 0    # сдвиг
alpha = 1   # масштаб
# создание распределения
our_laplace = laplace(beta, alpha)
```

```python
# создание интервала
number_of_steps = 1000
x = np.linspace(our_laplace.ppf(0.01), our_laplace.ppf(0.99), number_of_steps)
```

```python
default_dpi = plt.rcParamsDefault['figure.dpi']
plt.rcParams['figure.dpi'] = default_dpi*3
# построение функции вероятности и плотности функции распределения

ax_1 = plt.subplot(221)
ax_1.plot(x, our_laplace.cdf(x), 'b', lw=3, label='laplace pdf')
ax_1.set_title('Функция вероятности')
ax_1.set_xlabel('x')
ax_1.set_ylabel('f(x)')
ax_1.grid()

ax_2 = plt.subplot(222)
ax_2.plot(x, our_laplace.pdf(x), 'r-', lw=3, label='laplace pdf')
ax_2.set_title('Плотность распределения')
ax_2.set_xlabel('x')
ax_2.grid()

plt.show()
```

```python
# создание выборки из распределения Лапласа длиной 1000
r = our_laplace.rvs(size=1000)
```

```python
plt.rcParams['figure.dpi'] = default_dpi
plt.figure()
plt.hist(r, density=True, bins=50, color='b', label='средневыб. знач.')
plt.grid()
plt.plot(x, our_laplace.pdf(x), 'r', lw=4, label='плотность распр. Лапласа')
plt.legend(loc='best', frameon=True)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.show()
```

```python
# расчет параметров распределения Лапласа: матожидание и дисперсия
mean_l = beta
var_l = 2 / (alpha ** 2)
mean_l, var_l
```

```python
def norm_dist_plot(selection):
    avg = np.array([])
    number_of_selections = 1000
    for i in range(number_of_selections):
        l = our_laplace.rvs(size=selection)
        avg = np.append(avg, l.mean())
        
    # расчет параметров нормального распределения
    std = (var_l / selection)**0.5
    mean = mean_l
    
    print("Матожидание норм. распределения: {};\nСКО норм. распределения: {}.".format(mean, round(std, 3)))
    
    plt.figure()
    plt.hist(avg, bins=20, density=True, color='b', label='средневыб. знач.')
    plt.plot(np.sort(avg), norm.pdf(np.sort(avg), mean, std), 'r', lw=4, label='плотность норм. распр.')
    plt.grid()
    plt.title('Гистограмма выборочных средних из 1000 выборок с объемом n = {}'.format(selection))
    plt.legend(loc='best', frameon=True)
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.show()
```

```python
volume_of_selection = [5, 10, 30, 50, 100, 500]
```

```python
for volume in volume_of_selection:
    norm_dist_plot(volume)
```

# Выводы
1. На основе созданных выборок разного объема (от 5 до 500) была проверена работа центральной предельной теоремы;
2. При объеме выборки > 10 СКО нормального распределения меньше 0.5
3. При увеличении объема выборки нормальное распределение точнее аппроксимирует средневыборочные значения: дисперсия нормального распределения уменьшается.
